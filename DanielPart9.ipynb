```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.interpolate import interp1d
import numpy.random as npr

#Daniel's part 9
def pdf_z(theta,z):
    e = 1.6e-19
    #Electron and Muon mass squared
    mu2 = 105.658**2 #Mev^2
    me2 = 0.5110**2 #Mev^2
    me = 0.5110
    
    #Energy Decay####################################################
    X0 = 35.28 #cm
    E0 = 47571.03165490111
    s0 = 2*me*E0+2*me2#Jake gave me this equation for converting E to s
    s = s0 * math.exp(-z/X0) #Mev point 9, center beam energy
    #################################################################
    
    alpha = (e**2)/(4*math.pi)
    #constant to simplify calculation
    ce = 1-(4*me2/s)
    cu = 1-(4*mu2/s)
    #print(cu)
    c1 = ((alpha**2)/(4*s)) * (cu)**(1/2) * (ce)**(-1/2)
    
    #assuming me2 = 0
    a = ((alpha**2)/(4*s)) * math.sqrt(cu) * (1+ (4*mu2/s))
    b = ((alpha**2)/(4*s)) * cu
    sigma = (4*math.pi/3) *(3*a + b)
    return 2*math.pi*math.sin(theta)/sigma *(a + b*(math.cos(theta)**2))


def thetaGen_z(howMany,z):    
    thetas = []
    count = 0
    while count < howMany:
        x = npr.uniform(0,math.pi)
        y = npr.uniform(0,1)
        if y < pdf_z(x,z): 
            thetas.append(x)
            count += 1
    return thetas

#3D plot
#plots the PDF as a function of angle theta and block thickness z
res = 100
theta = np.linspace(0,math.pi,res)
z  = np.linspace(0,3,res)

T,Z = np.meshgrid(theta,z)
#print(T,Z)
PDF_z = np.empty((res,res))
for i in range(res):
    for j in range(res):
        PDF_z[i][j] = pdf_z(T[i][j],Z[i][j])

fig3Dpdf_z = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(T,Z,PDF_z,50,cmap = 'binary')
ax.set_xlabel('Theta')
ax.set_ylabel('z')
ax.set_zlabel('PDF');
##3dplot

#At the end of the block the energy should be such at the PDF is 0. But it stays high.
```
